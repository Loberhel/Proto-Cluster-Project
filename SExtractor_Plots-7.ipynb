```python
##Import the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Table column names
names = ['NUMBER', 'ELLIPTICITY', ' ERRTHETA_IMAGE ', 'DELTAPEAK_SKY', 'ALPHAPEAK_SKY', 'X_IMAGE', 'Y_IMAGE','FLUX_AUTO', 'FLUX_MODEL', 'MAG_AUTO', 'MAG_MODEL', 'MAGERR_AUTO', 'FLAGS', 'CLASS_STAR']
```

```python
"""
The first two cells make MAG_AUTO vs CLASS_STAR plots for optical filters u and g.
"""
```

```python
# g band filter catalog plots ADUs
## FWHM = 1.621 
#Importing the catalog into DataFrame
g8_ADU_catalog_data = pd.read_table('g8_filter_ADU_in.cat', sep = '\s+', header = None, skiprows = 14, names = names)
#print(g8_ADU_catalog_data)
#Remove the bright stars and galaxies after checking to make sure they were stars and close galaxies
g8_ADU_catalog_data_adjusted = g8_ADU_catalog_data[g8_ADU_catalog_data.MAG_AUTO >= 11.5] 
plt.plot(g8_ADU_catalog_data_adjusted['MAG_AUTO'],g8_ADU_catalog_data_adjusted['CLASS_STAR'],'.')
plt.xlabel('MAG_AUTO g')
plt.ylabel('CLASS_STAR')
plt.title("g-band MAG_AUTO vs CLASS_STAR [ADU]")
plt.show()
```

```python
u6_ADU_catalog_data_in = pd.read_table('u6_filter_ADU_in.cat', sep = '\s+', header = None, skiprows = 14, names = names)

#Remove the bright stars and galaxies after checking to make sure they were stars and close galaxies
u6_ADU_catalog_data_adjusted_in = u6_ADU_catalog_data_in[u6_ADU_catalog_data_in.MAG_AUTO >= 11.5] 
#print(u6_ADU_catalog_data_adjusted)
u6_ADU_catalog_data_adjusted_in = u6_ADU_catalog_data_adjusted_in[u6_ADU_catalog_data_adjusted_in.MAG_AUTO < 50] 
plt.plot(u6_ADU_catalog_data_adjusted_in['MAG_AUTO'],u6_ADU_catalog_data_adjusted_in['CLASS_STAR'],'.')
plt.xlabel('MAG_AUTO u')
plt.ylabel('CLASS_STAR')
plt.title("u-band MAG_AUTO vs CLASS_STAR [ADU]")

plt.show()
       
```

```python
"""
Combine the u and g catalogs along NUMBER
"""
```

```python
joined_u_g = u6_ADU_catalog_data_adjusted_in.join(g8_ADU_catalog_data_adjusted, on = 'NUMBER', lsuffix = '_u', rsuffix = '_g')
print(joined_u_g)
joined_u_g.to_csv(r'\Desktop\joined_u_g1.cat', header=True, index=None, sep='\t', mode='a')
```
