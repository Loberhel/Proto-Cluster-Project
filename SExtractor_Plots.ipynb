```python
##Import the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Table column names
names = ['NUMBER', 'DELTAPEAK_SKY', 'ALPHAPEAK_SKY', 'X_IMAGE', 'Y_IMAGE', 'XWIN_IMAGE', 'YWIN_IMAGE', 'XMODEL_IMAGE', 'YMODEL_IMAGE','FLUX_AUTO', 'FLUX_MODEL', 'MAG_AUTO', 'MAG_MODEL', 'MAGERR_AUTO', 'FLAGS',  'NITER_MODEL', 'CLASS_STAR']
```

```python
"""
The first two cells make MAG_AUTO vs CLASS_STAR plots for optical filters u and g.
"""
```

```python
# g band filter catalog plots ADUs
## FWHM = 1.621 
#Importing the catalog into DataFrame
g8_ADU_catalog_data = pd.read_table('g8_filter_ADU_in.cat', sep = '\s+', header = None, skiprows = 17, names = names)
#print(g8_ADU_catalog_data)
#Remove the bright stars and galaxies after checking to make sure they were stars and close galaxies
g8_ADU_catalog_data_adjusted = g8_ADU_catalog_data[g8_ADU_catalog_data.MAG_AUTO >= 11.5] 
plt.plot(g8_ADU_catalog_data_adjusted['MAG_AUTO'],g8_ADU_catalog_data_adjusted['CLASS_STAR'],'.')
plt.xlabel('MAG_AUTO g')
plt.ylabel('CLASS_STAR')
plt.title("g-band MAG_AUTO vs CLASS_STAR [ADU]")
plt.show()
```

```python
g8_ADU_catalog_data_in = pd.read_table('g8_filter_ADU_in.cat', sep = '\s+', header = None, skiprows = 17, names = names)
#Remove the bright stars and galaxies after checking to make sure they were stars and close galaxies with sigma = 5
g8_ADU_catalog_data_adjusted_in = g8_ADU_catalog_data_in[g8_ADU_catalog_data_in.MAG_AUTO >= 11.5] 
plt.plot(g8_ADU_catalog_data_adjusted_in['MAG_AUTO'],g8_ADU_catalog_data_adjusted_in['CLASS_STAR'],'.')
plt.xlabel('MAG_AUTO g')
plt.ylabel('CLASS_STAR')
plt.title("g-band MAG_AUTO vs CLASS_STAR [ADU]")
plt.show()
```

```python
#DEtermine what is deleted with the increase in magnitude limit

common = g8_ADU_catalog_data_adjusted.merge(g8_ADU_catalog_data_adjusted_in, on = ['CLASS_STAR'])
not_common = g8_ADU_catalog_data_adjusted[(~g8_ADU_catalog_data_adjusted.CLASS_STAR.isin(common.CLASS_STAR))&(~g8_ADU_catalog_data_adjusted.CLASS_STAR.isin(common.CLASS_STAR))]

plt.scatter(not_common['MAG_AUTO'], not_common['CLASS_STAR'])
plt.xlabel('MAG_AUTO')
plt.ylabel('CLASS_STAR')
plt.show()
```
